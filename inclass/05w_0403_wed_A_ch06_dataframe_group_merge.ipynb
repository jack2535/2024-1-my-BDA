```python
# 제조 회사 및 구동 방식별 분리
# city 평균 구하기
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean')) # 시험 예상
```

```python
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean')) # 시험 예상
```

```python
## audi의 drv별 빈도

# audi 추출
# drv별 분리
# 빈도 구하기
mpg.query('manufacturer == "audi"').groupby(['drv']).agg(n =('drv', 'count'))
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).count()
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).count()[['model']]
```

```python
mpg.groupby(['manufacturer', 'category']).agg(mean_cty = ('cty', 'mean'), mean_hwy = ('hwy', 'mean')) # 시험 예상
```

```python
mpg[['drv', 'cyl']].value_counts()
```

```python
mpg[['drv', 'cyl']].value_counts().to_frame()
```

```python
# drv 빈도 구하기
# 데이터 프레임으로 바꾸기, 변수명 n으로 바꾸기
# axis 이름에 drv 지정
# n이 100을 초과한 경우 추출
mpg['drv'].value_counts() \
          .to_frame('n') \
          .rename_axis('drv') \
          .query('n > 100')
```

```python
mpg['drv'].value_counts().to_frame('n').rename_axis('drv') # index의 제목 수정
```

```python
mpg.groupby('drv').agg(n = ('drv', 'count'))
```

```python
mpg.groupby('drv').agg(n = ('drv', 'count')).query('n > 100')
```

```python
# suv 추출
# 합산 연비 변수 만들기
# 제조 회사별로 분리
# 합산 연비 평균 구하기
# 내림차순 정렬
# 1~5위까지 출력, 시험 예상
mpg.query('category == "suv"') \
   .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
   .groupby('manufacturer') \
   .agg(mean_tot = ('total', 'mean')) \
   .sort_values('mean_tot', ascending = False) \
   .head()
```

##세로로 합치기: concat()

```python
# 학생 1~5번 시험 데이터 만들기
group_a = pd.DataFrame({'id'   : [1, 2, 3, 4, 5],
                        'test' : [60, 80, 70, 90, 85]})

# 학생 6~10번 시험 데이터 만들기
group_b = pd.DataFrame({'id'   : [6, 7, 8, 9, 10],
                        'test' : [70, 83, 65, 95, 80]})
```

```python
group_a  # group_a 출력
```

```python
group_b
```

```python
pd.concat([group_a, group_b])
```

```python
pd.concat([group_a, group_b]).reset_index()
```

```python
pd.concat([group_a, group_b]).reset_index(drop=True)
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
df1
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, how="right")
```

```python
df1.merge(df2, how="outer")
```

```python
df1.merge(df2, how="cross")
```
